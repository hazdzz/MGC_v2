```python
import pandas as pd
import numpy as np

import seaborn as sns
from plotnine import *
from plotnine.data import *
import matplotlib
import matplotlib.pyplot as plt
import palettable
#matplotlib.rcParams['text.usetex'] = True
```

```python
cornell_x = np.arange(195)
cornell_lap_y_1 = np.genfromtxt('cornell_sym_renorm_mag_lap_0_ev.csv', delimiter=',')
cornell_lap_y_2 = np.genfromtxt('cornell_sym_renorm_mag_lap_0.5_ev.csv', delimiter=',')
cornell_lap_y_3 = np.genfromtxt('cornell_sym_renorm_mag_lap_0.3333333333333333_ev.csv', delimiter=',')
cornell_lap_y_4 = np.genfromtxt('cornell_sym_renorm_mag_lap_0.25_ev.csv', delimiter=',')
cornell_lap_y_5 = np.genfromtxt('cornell_sym_renorm_mag_lap_0.2_ev.csv', delimiter=',')
```

```python
plt.figure(dpi=300)
sns.set_theme(style="white")

palette = sns.color_palette("viridis", 5)
fr = pd.DataFrame({'q=0':cornell_lap_y_1,
                   'q=1/2':cornell_lap_y_2,
                   'q=1/3':cornell_lap_y_3,
                   'q=1/4':cornell_lap_y_4,
                   'q=1/5':cornell_lap_y_5
                  })
p = sns.lineplot(data=fr, palette=palette)

plt.legend(loc='upper left')
plt.xlim(0, 195)
plt.ylim(0, 2)
plt.xlabel("Index")
plt.ylabel("Frequency response")
plt.savefig('FR_Cornell.png')
plt.show()
```
